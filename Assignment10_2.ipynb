```
# Kathryn Webb
# 4/8/2020
# 1: For the student table created last class, write a function that can add a student into the table
#   (1) The function accepts data in command line.
#   (2) Insert at least 10 students into the table.
#   (3) The major must be "data science", "information science", "computer science"
#   (4) Set at least 3 student in "data science", 4 in "information science", 5 in "computer science"
#
# 2: Write a function which can delete a student by stuid(in command line)
# 3: Write a function which can modify a student's name(in command line)
# 4: Write a function which can get the total count of students in one specific major(in command line)
# 5: Write a function which can get one student's id and name randomly(in command line)

import pymysql
import random as rnd
db = 'pydb'
pydb = pymysql.connect("localhost", "root", "Info3020!!!!!", db)
dbcursor = pydb.cursor()


def input_validate():
    u_input = input("\n 1: Enter new student" +
                    "\n 2: Delete existing student" +
                    "\n 3: Modify existing student information" +
                    "\n 4: Get total in one major" +
                    "\n 5: Get random student information" +
                    "\n 6: Exit" +
                    "\n Choice::: ")
    try:
        num_u_input = int(u_input)
    except ValueError:
        u_input = input("{} is not a number. Please select a choice between 1 and 6.".format(u_input))
    return int(u_input)


def insert_student():
    name = input("Name:")
    major = input("Major:")
    year = input("Year:")
    tuition = input("Tuition:")
    gpa = input("GPA:")
    values = (name, major, year, tuition, gpa)
    sql = "insert into tstudent(stuname, major, stuyear, tuition, GPA)" + "values(%s, %s, %s, %s, %s)"
    dbcursor.execute(sql, values)
    pydb.commit()
    dbcursor.close()
    pydb.close()


def delete_student():
    sql = "delete from tstudent where stuid=%s"
    stuid = input("Enter student ID for deletion:")
    dbcursor.execute(sql, stuid)
    pydb.commit()
    dbcursor.close()
    pydb.close()


def modify_student():
    stuid = input("Enter ID of student:")
    update_col = input("Enter column name:")
    update_info = input("Enter new information:")
    sql = "update tstudent set {0}=%s where stuid=%s".format(update_col)
    values = (update_info, stuid)
    dbcursor.execute(sql, values)
    pydb.commit()
    dbcursor.close()
    pydb.close()


def total_per_major():
    major = input("Which major:")
    sql = "select count(*) from tstudent where major=%s"
    dbcursor.execute(sql, major)
    results = dbcursor.fetchone()
    print(results[0])
    dbcursor.close()
    pydb.close()


def random_student():
    sql = "select max(stuid) from tstudent"
    dbcursor.execute(sql)
    maxid = dbcursor.fetchone()
    sql = "select min(stuid) from tstudent"
    dbcursor.execute(sql)
    minid = dbcursor.fetchone()
    sql = "select * from tstudent where stuid=%s"
    stuid = rnd.randint(minid[0], maxid[0])
    dbcursor.execute(sql, stuid)
    print(dbcursor.fetchone())
    dbcursor.close()
    pydb.close()


user_input = input_validate()
if user_input == 1:
    insert_student()
elif user_input == 2:
    delete_student()
elif user_input == 3:
    modify_student()
elif user_input == 4:
    total_per_major()
elif user_input == 5:
    random_student()
elif user_input == 6:
    print("Exit")
```
